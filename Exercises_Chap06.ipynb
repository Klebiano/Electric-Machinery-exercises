```python
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.print_figure_kwargs={'facecolor' : "w"}
plt.rcParams.update({'font.size': 10, 'font.weight':'bold', 'figure.figsize': (8,6)})
#props={'weight':'bold'}                                                                
np.seterr(divide='ignore', invalid='ignore');
```

```python
#Exercise 6.4

V1 = 460/np.sqrt(3)
Nf = 3
polos = 4
fe = 60
R1 = 0.163
X1 = 0.793j
X2 = 1.101j
Xm = 18.87j
```

```python
omega = 4*np.pi*fe/polos
ns = 120*fe/polos
```

```python
Z1eq = Xm*(R1+X1)/(R1 + (X1+Xm));
R1eq = np.real(Z1eq);
X1eq = np.imag(Z1eq);
V1eq = abs(V1*Xm/(R1 + (X1+Xm)))
```

```python
R2 = [0.1,0.2,0.5,1.0,1.5]

s = np.linspace(0,1,1000)
rpm = ns*(1-s)

#linha = ['.-','--','*-', '-+','.']

for i in range(len(R2)):
    I2 = abs(V1eq/(Z1eq + X2 + R2[i]/s))
    Tmech = Nf*(I2**2)*R2[i]/(s*omega)
    plt.plot(rpm,Tmech,linewidth=2)


plt.xlabel('rpm', fontweight='bold')
plt.ylabel('Torque (N.m)', fontweight='bold')
plt.xlim(0,1800)
plt.ylim(0,270)
plt.xticks(np.arange(0, 1801, step=180))
plt.yticks(np.arange(30, 271, step=30))
plt.legend(R2, loc='best', title='$\\bf{R2}$', ncol=len(R2))
plt.title('Torque X rotao', fontweight='bold')
plt.show()
```
